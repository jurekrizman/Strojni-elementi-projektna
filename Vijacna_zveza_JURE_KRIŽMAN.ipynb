## Knjižnica vijakov

## Preliminarna določitev dimenzij vijačne zveze

```python
dt = D + 2*e + 2*s # mm premer, na katerem so nameščeni vijaki
dt
```

```python
zd = (np.pi * dt)/(h+dw) # (število vijakov)
z = round(zd)
if z < zd:
    z += 1
z
```

```python
Fb = (pmax*100000) * (np.pi * (D/1000)**2 / 4) # N (skupna delovna sila)
Fb
```

```python
As = 1.23 * ka * (Fbi+Fkli) / (Nap_dop) # mm^2 (minimalni prerez vijaka)
As
```

# Določitev podajnosti elementov vijačne zveze

## Legenda 4.poglavje

A1 - prerez stebla vijaka in imenski prerez vijaka 

A3 - prerez jedra navoja


d - imenski nosilni prerez (od zgori na drugo stran) - v Prebilu d 

d2 - srednji premer navoja (od uspodi na drugo stran) - v Prebilu d1 

d3 - jedrni premer navoja (od sredine navoja do druge strani) - v Prebilu d2


l1 - dolžina stebla 

l3 - dolžina navoja 


cs0 - togost glave vijaka

cs1 - togost stebla vijaka

cs3 - togost navoja vijakaž

cs4 - togost matice

csi - array vsega

cs - skupna togost vijačne zveze


dsi - array podajnosti vsakega dela vijačne zveze 

ds - skupna podajnost vijačne zveze

## Podajnost vijaka

```python
f = (Fbi*l)/(An*Es) # mm (raztezek)
f
```

```python
AN = np.pi* d1**2 /4 # mm^2 (imenski prerez vijaka)

A1 = np.pi* d1**2 /4 # mm^2 (1. prerez stebla vijaka)
l1 = l-(2*d1 +12) # mm (1. dolžina stebla vijaka)

A2 = An # mm^2 (nosilni prerez navoja)
l2 = 2*h - l1 # mm (dolžina neovitega navoja)

A3 = np.pi* d3**2 /4 # mm^2 (prerez jedra navoja)
l3 = 2*d1 +12 # mm (dolžina navoja)
```

```python
cs0 = AN*Es/(d1*.4) # N/mm (togost glave vijaka)
cs1 = A1*Es/l1 # N/mm (togost 1. stebla vijaka)
cs2 = A2 * Es / l2 # N/mm (togost neovitega navoja)
cs3 = A3*Es/(.5*d1) # N/mm (togost navoja vijaka) ?? verjetno dolžina navoja matice
cs4 = AN*Es/(d1*.4) # N/mm (togost matice)

csi = np.array([cs0,cs1,cs2,cs3,cs4])
csi
```

```python
cs = (sum(csi**(-1)))**(-1) # N/mm (skupna togost vijačne zveze)
cs
```

```python
dsi = csi**(-1) # mm/N (elastičnost (podajnost) posameznega dela vijačne zveze)
dsi
```

```python
ds = sum(dsi) # mm/N (elastičnost (podajnost) vijačne zveze)
ds
```

## Podajnost spenjalnih delov

## Legenda za podajnost

dw - zev ključa

Da = tbp - premer tlačnega stožca

t - da se prekrivajo 

dt - podajnost spenjalnih delov


```python
#Da < dw
Da = dw +2* h * np.sin(fi)
Da1 = dw + h
dh = d1 + 1
lk = 2*h
t = Da - 0.1*Da # neka varnost za prekrivanje v tem primeru 10%
print(dh)
print(Da)
```

```python
dt = ((2/(w*dh*np.tan(fi)))*np.log(((dw+dh)*(Da-dh))/((dw-dh)*(Da+dh)))+((4/(Da**2-dh**2))*(lk-((Da-dw)/(w*np.tan(fi))))))/(Et*np.pi)
dt #Podajnost spenjalnih delov, to je za delno razvit
```

# Moment privitja vijaka

## Legenda 6.poglavje

P - korak navoja 

μg - koeficient trenja med navoji 

ϕ - Kot vzpona vijačnice

## Torne razmere na navojih


```python
ϕ = np.arctan(P/(np.pi*d2)) #Kot vzpona vijačnice
ϕ #rad
```

```python
ρ = np.arctan(μg/(np.cos(β/2)))
ρ #rad
#Torni kot za vijake s trikotnim in trapeznim profilom
```

## Moment privitja


## Legenda za moment privitja
Mg - moment privijanja

Mra - moment odvijanja

Ma - moment privitja vijaka

μk - koeficient trenja pod matico oz. glavo vijaka 

Fv = Fvm - sila prednapetja

Frg - obodna sila na navoju

dk - srednji premer kolobarja pod matico


```python
Frg = Fv*np.tan(ϕ+ρ)
Frg #N
```

```python
Mg = Frg*(d2/2)
Mg #Nmm
```

```python
dk = (dw+dh)/2
dk #mm
```

```python
Mra = Fv*μk*(dk/2)
Mra #Nmm
```

```python
Ma = Mg + Mra
Ma #Nmm
```
